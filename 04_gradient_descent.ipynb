```python
#importing packages
import numpy as np
```

```python
def gradient_descent(x,y):
    m_curr = b_curr = 0
    iterations = 10000
    n = len(x)
    learning_rate = 0.001

    for i in range(iterations):
        y_predicted = m_curr * x + b_curr                        # General Equation: y = mx + b
        cost = (1/n) * sum([val**2 for val in (y-y_predicted)])  # Mean Squared Error  
        md = -(2/n)*sum(x*(y-y_predicted))                       # Partial Derivative of m
        bd = -(2/n)*sum(y-y_predicted)                           # Partial Derivative of b
        m_curr = m_curr - learning_rate * md
        b_curr = b_curr - learning_rate * bd
        print ("m: {:8f}, b: {:8f}, cost: {:8f}, iteration: {}".format(m_curr,b_curr,cost, i))
        
#When m_curr and b_curr reaches the optimum value,the cost function becomes constant on further iterations.(for iterations >970)
```

```python
x = np.array([1,2,3,4,5])
y = np.array([5,7,9,11,13])

gradient_descent(x,y)
```
